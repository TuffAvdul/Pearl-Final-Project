```python
import pandas as pd
import numpy as np
import copy
```

### import data

```python
business = pd.read_json(r"yelp_academic_dataset_business.json" , lines = True)
review = pd.read_json(r"yelp_academic_dataset_review.json" , lines = True)
```

### take copy of the data

```python
bsn = copy.deepcopy(business)
rw = copy.deepcopy(review)
```

```python
class data_cleaning:
    
    #class attributes
    remove_zero_index = [] # to remove the index of zero category
    category_set = set() # set of all categories
    category_list= [] # make it list to organize it
    
    kw_index_list = [] # index of specified kw search
    
    data_log = dict() # every dataset ["name" : index_value]
    review_data_log = dict() # review data ["name" , index_value]
    
    #constructor
    def __init__(self , data , review_data , column):
        self.data = data
        self.column = column
        self.review_data = review_data
        
        for i in range(len(self.data)):
            filt = data.loc[i,str(column)]
            for j in range(len(filt)):
                data_cleaning.category_set.add(str(j))
        
        data_cleaning.category_list = list(data_cleaning.category_set)
    
    
    # function to remove the zero category in the data
    # function 0.0
    def error_zero(self):
        error_category_zero = dict()

        for i in range(len(self.data)):
            category = self.data.loc[i,"categories"]
            if (len(category) >= 1):
                pass
            else:
                error_category_zero[i] = self.data.loc[i,["business_id" , "name"]]
        data_cleaning.remove_zero_index = (pd.DataFrame(error_category_zero).T).index
    
    # function 0.1 / remove the zero index
    def remove_error_zero(self):
        self.data = self.data.drop(data_cleaning.remove_zero_index)
        self.data = self.data.reset_index(drop=True)
        print(f"the no. columns of the data is {self.data.shape[0]}")
        
    
    # function that return index of specific data with keywords
    # function 1
    def get_category(self , keywords):
        
        for i in range(len(self.data)):
            check_list = self.data.loc[i,str(self.column)]
            for j in range(len(check_list)):
                if check_list[j] in keywords:
                    data_cleaning.kw_index_list.append(i)
                    break
                    
    # function to seperate the data according to the index
    #function 2 
    def seperate_category(self , new_data_name):
        # saving the data to a dictionary to retrive it 
        data_cleaning.data_log[new_data_name] = data_cleaning.kw_index_list
        
        # Separating the review_business_data according to the index_list
        S_list_index = data_cleaning.data_log[new_data_name]
        business_list = self.data.loc[S_list_index,:]
        bus_list2 = list(business_list.business_id)

        rw_list_index = []
        for i in range(len(self.review_data)):
            filt = self.review_data.loc[i, "business_id"]
            if str(filt) in bus_list2:
                rw_list_index.append(self.review_data.index[i])

        data_cleaning.review_data_log[new_data_name] = rw_list_index
        return business_list
```
